```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import (
    make_moons
)
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Dữ liệu hình bán nguyệt
X, _ = make_moons(
    n_samples=200,
    noise=0.05,
    random_state=42
)
```

```python
# Khởi tạo
dbscan = DBSCAN(
    eps=0.3,
    min_samples=5
)

# Fit và predict
labels = dbscan.fit_predict(X)

# Số cụm (không tính noise -1)
n_clusters = len(
    set(labels)
) - (1 if -1 in labels else 0)
print(f"Clusters: {n_clusters}")
```

```python
# Vẽ kết quả
plt.figure(figsize=(10, 6))

# Tách noise và cụm
unique_labels = set(labels)
colors = plt.cm.Spectral(
 np.linspace(0, 1, 
 len(unique_labels))
)

for k, col in zip(
 unique_labels, colors
):
 if k == -1:
    col = [0, 0, 0, 1] 
 # Đen cho noise
 
 mask = (labels == k)
 plt.scatter(
 X[mask, 0],
 X[mask, 1],
 c=[col],
 marker='o' if k != -1 
 else 'x'
 )

plt.title('DBSCAN Clustering')
plt.show()
```
