```python
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
# Tạo dữ liệu mẫu
X, _ = make_blobs(
    n_samples=50,
    centers=3,
    random_state=42
)
```

```python
# Tính linkage matrix
Z = linkage(X, method='ward')
```

```python
# Vẽ dendrogram
plt.figure(figsize=(10, 6))
dendrogram(Z)
plt.title('Dendrogram')
plt.xlabel('Data Point Index')
plt.ylabel('Distance')
plt.show()
```

```python
# Khởi tạo model
agg_clust = AgglomerativeClustering(
    n_clusters=3,
    linkage='ward'
)

# Fit và predict
labels = agg_clust.fit_predict(X)

# Vẽ kết quả
plt.scatter(
    X[:, 0], X[:, 1],
    c=labels,
    cmap='viridis'
)
plt.title('Agglomerative Clusters')
plt.show()
```

```python
linkages = [
 'ward', 'complete',
 'average', 'single'
]

fig, axes = plt.subplots(
 2, 2, figsize=(12, 10)
)

for i, method in enumerate(linkages):
 agg = AgglomerativeClustering(
 n_clusters=3,
 linkage=method
 )
 labels = agg.fit_predict(X)

 ax = axes[i//2, i%2]
 ax.scatter(
 X[:, 0], X[:, 1],
 c=labels
 )
 ax.set_title(method)
```
